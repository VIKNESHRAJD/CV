```python
import cv2
import numpy as np
```

```python
img = cv2.imread('A2.jpeg')
img
```

```python
color = cv2.bilateralFilter(img, 9, 275, 275)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
gray = cv2.medianBlur(gray, 5)
```

```python
th = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)

cartoon = cv2.bitwise_and(color, color, mask=th)

cv2.imshow("Original", img)
cv2.imshow("Cartoon Effect", cartoon)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
