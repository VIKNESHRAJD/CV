# IMPORT LIBRARY

```python
import cv2
import numpy as np
```

# Read Image - Source

```python
img = cv2.imread('SQ.jpg')
```

# Gray Conversion

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

# Read Image - Template

```python
temp = cv2.imread('SQ1.jpg',0)
```

# Display Image

```python
cv2.imshow('Source',img)
cv2.imshow('Template',temp)

if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

# Template width and height

```python
print(temp.shape)
```

```python
w,h = temp.shape[::-1]
```

```python
w
```

```python
h
```

# Template Matching

```python
mat = cv2.matchTemplate(img_gray,temp,cv2.TM_CCOEFF_NORMED)
```

# Setting Thrshold

```python
threshold = 0.7
```

# Matched Location

```python
location = np.where(mat >=threshold)
```

# Matched Region

```python
for pt in zip(*location[::-1]):cv2.rectangle(img,pt,(pt[0] + w, pt[1] + h),(0,255,255),2)
```

# Output Display

```python
cv2.imshow("Detected",img) 
if cv2.waitKey(1) & 0xFF == 27:
    cv2.destroyAllWindows()

```
